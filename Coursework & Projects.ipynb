### Business Analytics Final Project, Brooklyn College, Tableau, 2021

Group project of 6. 
Health Analysis of 11 countries across a time span of 12 years with 4 categories containing 17 variables. 
Countries are grouped by region and health variables are cross-examined with Social Protection and Labor Force, Economic and Population Variables. 
Each member focused on a hypothesis that interested them while giving feedback and helping with other hypotheses. 
Worked on Hypothesis 1 which investigated a possible correlation of Unemployment(% of total labor force) to Domestic private health expenditure (% of current health expenditure). 
A vast number of our countries had universal healthcare systems that did not rely on employment status but did offer caps or lower premiums for the unemployed.
